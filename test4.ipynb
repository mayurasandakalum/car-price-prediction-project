```python
import json
```

```python
with open('server/app/json_data/label_encodings.json') as f:
    data = json.load(f)
```

```python
data
```

```python
def recursive_lower_dict(input_dict):
    if isinstance(input_dict, dict):
        return {key.lower(): recursive_lower_dict(value) if isinstance(value, dict) else value.lower() if isinstance(value, str) else value
                for key, value in input_dict.items()}
    elif isinstance(input_dict, list):
        return [recursive_lower_dict(item) for item in input_dict]
    else:
        return input_dict
    
data2 = recursive_lower_dict(data)
```

```python
with open('server/app/json_data/label_encodings_2.json', 'w') as json_file:
    json.dump(data2, json_file, indent=2)

print("JSON file 'output.json' has been saved.")
```
